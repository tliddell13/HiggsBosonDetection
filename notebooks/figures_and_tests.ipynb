```python
#Imports
import pandas as pd
import numpy as np
import torch
import pickle
import matplotlib.pyplot as plt
```

## Figures

```python
# Figure 

# Training the larger models without dropout caused the model to overfit the training data as ssen in the training and testing losses
# Load the training and testing losses for the model with no dropout
NoDoModel4_train_losses = pickle.load(open("../training_outputs/noDO_train_losses.pkl", "rb"))
NoDoModel4_test_losses = pickle.load(open("../training_outputs/noDO_test_losses.pkl", "rb"))

# Plot the training and testing losses for the model with no dropout
plt.figure()
plt.plot(NoDoModel4_train_losses, label='No Dropout Train')
plt.plot(NoDoModel4_test_losses, label='No Dropout Test')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Testing Losses for Model 4 with No Dropout')
plt.legend()
plt.show()


# Training and testing losses for the different MLP models as compared to number of epochs
# Load the training and testing losses for the different MLP models
model1_train_losses = pickle.load(open("../training_outputs/model1_train_losses.pkl", "rb"))
model1_test_losses = pickle.load(open("../training_outputs/model1_test_losses.pkl", "rb"))
model2_train_losses = pickle.load(open("../training_outputs/model2_train_losses.pkl", "rb"))
model2_test_losses = pickle.load(open("../training_outputs/model2_test_losses.pkl", "rb"))
model3_train_losses = pickle.load(open("../training_outputs/model3_train_losses.pkl", "rb"))
model3_test_losses = pickle.load(open("../training_outputs/model3_test_losses.pkl", "rb"))
model4_train_losses = pickle.load(open("../training_outputs/model4_train_losses.pkl", "rb"))
model4_test_losses = pickle.load(open("../training_outputs/model4_test_losses.pkl", "rb"))
model5_train_losses = pickle.load(open("../training_outputs/model5_train_losses.pkl", "rb"))
model5_test_losses = pickle.load(open("../training_outputs/model5_test_losses.pkl", "rb"))

# Plot the model4 training and testing losses for the firsst 1200 epochs
plt.figure()
plt.plot(model4_train_losses[:1200], label='Model 4 Train')
plt.plot(model4_test_losses[:1200], label='Model 4 Test')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Testing Losses for Model 4')
plt.legend()
plt.show()

# Plot the training losses for the different MLP models
plt.figure()
plt.plot(model1_train_losses, label='Model 1 Train')
plt.plot(model2_train_losses, label='Model 2 Train')
plt.plot(model3_train_losses, label='Model 3 Train')
plt.plot(model4_train_losses, label='Model 4 Train')
plt.plot(model5_train_losses, label='Model 5 Train')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Losses for the Different MLP Models')
plt.legend()
plt.show()

# Plot the testing losses for the different MLP models
plt.figure()
plt.plot(model1_test_losses, label='Model 1 Test')  
plt.plot(model2_test_losses, label='Model 2 Test')
plt.plot(model3_test_losses, label='Model 3 Test')
plt.plot(model4_test_losses, label='Model 4 Test')
plt.plot(model5_test_losses, label='Model 5 Test')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Testing Losses for the Different MLP Models')
plt.legend()
plt.show()

# Model 1 
"""
Training time: 452.27327489852905
F1: 0.7686924317162332
Accuracy: 0.7501758620689655
Confusion Matrix: [[ 97168  39523]
 [ 32926 120383]]
"""

# Model 2
"""
Early stopping at epoch 1827
Training time: 570.1542136669159
F1: 0.7772593272290662
Accuracy: 0.763951724137931
Confusion Matrix: [[102110  34581]
 [ 33873 119436]]
"""

# Model 3 
"""
Training time: 1253.3577842712402
F1: 0.789177397790444
Accuracy: 0.7727172413793103
Confusion Matrix: [[100723  35968]
 [ 29944 123365]]
"""

# Model 4
"""
Training time: 2297.4291875362396
F1: 0.7951814069686015
Accuracy: 0.7760965517241379
Confusion Matrix: [[ 99023  37668]
 [ 27264 126045]]
"""
```
